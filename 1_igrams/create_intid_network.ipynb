```python
import os,sys
import glob
#import pandas as pd
import math
#import scipy
#from osgeo import gdal
import numpy as np
from array import array
#from scipy  import signal
import matplotlib.pyplot as plt
#import matplotlib.patches as mpatches
#from scipy.stats import gaussian_kde
import itertools
%matplotlib  widget
```

```python
# load in dates
workdir = '/data/ocp4/LagunaSalada/Data_and_Figs/cropped_9000_1800_18000_3600/'
dates = [os.path.basename(x) for x in glob.glob(workdir+"SLC_vv/2*")]
dates = sorted(dates)
nd = len(dates); numIgrams = nd-1
```

```python
# How many potential igram pairs are there in my 211 dates? Spanning all different lengths
#num_potential_igrams = scipy.special.comb(nd,2,exact=True)
num_potential_igrams = math.comb(nd,2)
print(str(num_potential_igrams)+' potential pairs.')
```

```python
def nchoosek(startnum, endnum, step=1, n=2):
    num_potential_igrams = math.comb(endnum-startnum,2)+(endnum-startnum)
    c = np.zeros((num_potential_igrams,2))
    count = 0
    for i in itertools.combinations(range(startnum,endnum+1,step),n):
        c[count,0] = i[0]
        c[count,1] = i[1]
        count = count+1
    return c
intid = nchoosek(0, nd-1)
print(intid)
```

```python
diffs = np.diff(intid,axis=1)
for i in range(np.shape(intid)[0]):
    print(intid[i],diffs[i])
```

```python
inds = np.where(diffs <= 3)
#print(np.shape(inds))
#print(inds[0])
```

```python
intid2_a = intid[inds,0]
intid2_b = intid[inds,1]
intid2_a = intid2_a[0]
intid2_b = intid2_b[0]
intid2 = np.vstack((intid2_a,intid2_b))
intid2 = intid2.T
print(np.shape(intid2))
print(intid2)
```

```python
intid3 = np.ndarray((np.shape(intid2)),dtype='int')
for i in range(np.shape(intid2)[0]):
    intid3[i,0] = int(intid2[i,0])
    intid3[i,1] = int(intid2[i,1])
print(intid3)
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Fixing random state for reproducibility
np.random.seed(19680801)

# Compute pie slices
N = 20
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
radii = 10 * np.random.rand(N)
width = np.pi / 4 * np.random.rand(N)
colors = plt.cm.viridis(radii / 10.)

angles = 2*np.pi*np.random.rand(500)

counts = np.histogram(angles,theta)

ax = plt.subplot(projection='polar')
ax.bar(theta, radii, width=width, bottom=0.0, color=colors, alpha=0.5)
#ax.bar(theta, radii, width=width, color=colors, alpha=0.5)

ax.plot(2,2,'o')
plt.show()

ax = plt.subplot()
for i in range(N-1):
    x=np.array([0,counts[0][i]*np.cos(theta[i]),counts[0][i]*np.cos(theta[i+1]),0])
    y=np.array([0,counts[0][i]*np.sin(theta[i]),counts[0][i]*np.sin(theta[i+1]),0])
    ax.plot(x,y,'r')
    #ax.plot(theta, radii, width=width, bottom=0.0, color=colors, alpha=0.5)

x=np.array([0,counts[0][-1]*np.cos(theta[-1]),counts[0][-1]*np.cos(theta[0]),0])
y=np.array([0,counts[0][-1]*np.sin(theta[-1]),counts[0][-1]*np.sin(theta[0]),0])
ax.plot(x,y,'r')

ax.plot(2,2,'o')
plt.axis('image')
plt.show()
```

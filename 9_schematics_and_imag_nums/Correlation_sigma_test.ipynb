```python
# this is translated from Rowena's MATLAB script
import os,sys
import numpy as np
import matplotlib.pyplot as plt
%matplotlib  widget
```

```python
# number of trials
n = int(1e3)

# correlations between 0 and 1
cors = np.linspace(0.05,1,50)
# standard deviations using the formula 
sigs = np.sqrt(-2*np.log(cors))


# initialize empty array for ??
allmeanmags = np.zeros(len(cors))

for i in np.arange(0,len(cors)):
    sig = sigs[i]
    # generate noise (zero phase mean)
    phs = np.random.randn(1,n)*sig
    data = np.exp(1j*phs)
    # mean phase error 
    meandat = np.mean(data)
    meanmag = np.abs(meandat)
    
    allmeanmags[i] = meanmag
```

```python
# plot it
plt.figure(figsize=(7,3.5), tight_layout=True)
plt.subplot(1,2,1)
plt.plot(cors,sigs)
plt.xlabel('correlation')
plt.ylabel('sigma')

plt.subplot(1,2,2)
plt.plot(cors,cors,'k')
plt.plot(cors,allmeanmags,'.-')
plt.xlabel('correlation')
plt.ylabel('calculated correlation w/ given sigma')

plt.show()
```

```python
# practice with imaginary numbers
import os,sys
import numpy as np
import matplotlib.pyplot as plt
%matplotlib  widget
```

```python
# create two imaginary numbers 
a = 4+3*1j
a2 = np.sqrt(np.real(a)*np.real(a) + np.imag(a)*np.imag(a))*np.exp(1j*(np.arctan(np.imag(a)/ \
                                                                                 np.real(a)))) 
b = 7.5+10*1j
b2 = np.sqrt(np.real(b)*np.real(b) + np.imag(b)*np.imag(b))*np.exp(1j*(np.arctan(np.imag(b)/ \
                                                                                 np.real(b)))) 

# a=a2 and b=b2
```

```python
# plot it
ax=plt.figure(figsize=(4,4), tight_layout=True)
plt.scatter(np.real(a),np.imag(a))
plt.scatter(np.real(a2),np.imag(a2))
plt.scatter(np.real(b),np.imag(b))
circle=plt.Circle([0,0],1,fill=False)
plt.xlabel('real')
plt.ylabel('imag')
plt.grid()
plt.gca().add_patch(circle)
plt.show()
```

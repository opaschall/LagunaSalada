```python
t1 = np.arange(0,365.25,12)
t2 = np.arange(365.25,365.25*2,6)
t = np.hstack((t1,t2))
```

```python
disp = 0.01*t
disp2 = 0.011*t
disp3 = 0.0

t2 = t/365.25

plt.figure(figsize=(5,3),layout='constrained')
plt.plot(t2,disp,c=[0.9290, 0.6940, 0.1250],linewidth=2.5)
plt.plot(t2,disp2,c='tab:red',ls='--',linewidth=2.5)
plt.xlabel('time (yrs)')
plt.ylabel('displacement (cm)')
plt.xlim(t2.min(),t2.max()); plt.ylim(disp2.min(),disp2.max())
plt.show()
```

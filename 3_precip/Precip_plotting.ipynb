```python
#import os,sys
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
%matplotlib  widget
```

```python
# Read in raw precipitation data and plot it 
file = '/data/ocp4/LagunaSalada/Data_and_Figs/3_precip/precip.txt'

precip = np.loadtxt(file, delimiter=',', skiprows=3, dtype=str)

precipDates = precip[:,0]
precip = precip[:,1]
#print(type(precip[0]))
precip = precip.astype(float)
#print(precip)
# convert to cm from inches 
precip = precip*2.54

# make the dates from precip.txt into datetime objects
my_dates=[]
for i in range(0,len(precipDates)):
    my_dates.append(datetime.strptime(str(precipDates[i]).replace('\n',''), '%Y-%m-%d'))

fig=plt.figure(figsize=(12,3), tight_layout=True)
#plt.xticks(np.arange(0, np.size(precipDates)+1, 182))
plt.xticks(rotation = 70)

#plt.plot(precipDates, precip,'-m')
plt.plot(my_dates, precip,'-m')
plt.xlabel('date')
plt.ylabel('precipitation (cm)')
plt.xlim([np.min(my_dates), np.max(my_dates)])
#plt.xlim([0, np.size(precip)])
plt.ylim([0,0.5])
plt.title('Precipitation in El Centro, CA')
plt.show()
```

```python
# NONZERO precip values with their dates
print('date \t\t       precip\n')
for i in np.arange(len(precip)): 
    if precip[i] > 0.0:
        print(str(my_dates[i])+'    '+str(precip[i])+' cm')
```

```python
# BIGGER STORMS
# days with precip > 0.5cm 
print('date \t\t       precip > 0.5cm\n')
for i in np.arange(len(precip)): 
    if precip[i] > 0.5:
        print(str(my_dates[i])+'    '+str(precip[i])+' cm')
```
